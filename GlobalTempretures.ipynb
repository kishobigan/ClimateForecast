```python
df['dt'] = pd.to_datetime(df['dt'])
print(df['dt'].min())
print(df['dt'].max())
```

```python
df.set_index('dt').plot(kind='line', legend=False, title='Global Tempratures')
```

```python
plt.fill_between(df['dt'],
                 df['LandAverageTemperature'] - df['LandAverageTemperatureUncertainty'],
                 df['LandAverageTemperature'] + df['LandAverageTemperatureUncertainty'],
                 alpha=0.2, label='Uncertainty', color='red')
plt.legend()
```

```python
df[['LandMaxTemperature', 'LandMinTemperature', 'LandAverageTemperature']].plot()
```

```python
df['RollingAvg'] = df['LandAverageTemperature'].rolling(window=12).mean()
df['RollingAvg'].plot(title='12-Month Rolling Average of Land Average Temperature')
```

```python
df['year'] = df['dt'].dt.year
yearly_average = df.groupby('year')['LandAverageTemperature'].mean()
yearly_average.plot(kind='line', legend=False, title='Yearly Average Temperature')
```

```python
df.iloc[-1]
```

```python
last_year = df['year'].max()

# Filter data for the last year
last_year_data = df[df['dt'].dt.year == last_year]

# Plot
plt.plot(last_year_data['dt'], last_year_data['LandAverageTemperature'], marker='o')
plt.title(f"Data for the Year {last_year}")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid()
plt.show()
```

```python
last_year = df['year'].max()

last_year_data = df[df['dt'].dt.year == last_year]
last_before_year_data = df[df['dt'].dt.year == last_year-1]

# Plot
plt.plot(last_year_data['dt'], last_year_data['LandAverageTemperature'], marker='o')
plt.plot(last_before_year_data['dt'], last_before_year_data['LandAverageTemperature'], marker='s')
plt.title(f"Data for the Year {last_year}")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid()
plt.show()
```

```python
year_2000_2015_data = df[df['year'].isin([year for year in range(2000, 2016)])].copy()

year_2000_2015_data['month'] = year_2000_2015_data['dt'].dt.month

monthly_avg_temp = year_2000_2015_data.groupby(['year', 'month'])['LandAverageTemperature'].mean().unstack(level=0)

plt.figure(figsize=(12, 8))
monthly_avg_temp.plot(marker='o')
plt.title(f"Monthly average temperature for 2000 to 2015")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.xticks(ticks=range(1, 13), labels=["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
plt.legend(title="year")
plt.grid()
plt.show()
```

```python
monthly_avg_temp_for_all_years = year_2000_2015_data.groupby(['month', 'year'])['LandAverageTemperature'].mean().unstack(level=0)

plt.figure(figsize=(12, 8))
plt.plot(monthly_avg_temp_for_all_years.mean(), marker='s', color='red')
plt.plot(monthly_avg_temp_for_all_years.min(), marker='o', color='blue')
plt.plot(monthly_avg_temp_for_all_years.max(), marker='o', color='green')
plt.title(f"Monthly average tempreture for all years")
plt.xlabel("Month")
plt.ylabel("tempreture")
plt.grid()
plt.show()
```

```python
monthly_avg_temp_for_all_years.plot(kind='box', legend=False, title='Monthly average tempreture for all years')
```

```python
df['month'] = df['dt'].dt.month
data_grouped_by_month = df.groupby(['month', 'year'])['LandAverageTemperature'].mean().unstack(level=0)
plt.figure(figsize=(12, 8))
data_grouped_by_month.mean().plot(kind='line', legend=False, title='Monthly average tempreture', marker='o')
plt.grid(True)
plt.xticks(range(12))
plt.show()
```
